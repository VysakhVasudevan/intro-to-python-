```python
fifadataframe= pandas.read_csv("players_20.csv")
min_age=fifadataframe.age.min()
min_age_loc = fifadataframe.age.idxmin()
person=fifadataframe.loc[min_age_loc]

print(person.long_name," has the lowest age of ",min_age)
```
